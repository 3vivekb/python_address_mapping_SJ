```python
import folium

# Get a basic world map.
homes_map = folium.Map(location=[37.30, -121.9], zoom_start=10)
```

```python
# Draw markers on the map.
counter = 0
record = 0
for what, row in homes.iterrows():
    if not pd.isnull(row["STREET"]):
        #Make points with street names GREEN
        folium.Marker(location=[row["LAT"], row["LON"]], popup=row["ID"], icon=folium.Icon(color='green')).add_to(homes_map)

    
    else:
        #Make points without street names RED
        folium.Marker(location=[row["LAT"], row["LON"]], popup=row["ID"], icon=folium.Icon(color='red')).add_to(homes_map)

    record += 1

    counter += 1
    if(counter == 500):
        break
# Create and show the map.
```
